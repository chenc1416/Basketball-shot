```python
import numpy as np
import pandas as pd
import re

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get_data(df):
    data = ''
    for n in range (0, len(df)):
        term = ''.join(str(df.loc[n]))
        data += term
    
    return data
```

```python
df1 = pd.read_excel('/Users/chen/Desktop/DATA-1.xlsx')  
df2 = pd.read_excel('/Users/chen/Desktop/DATA-2.xlsx')

data1 = get_data(df1)
data2 = get_data(df2)
```

```python
def data_tokenize(data):
    
    df = pd.DataFrame()
        
    TimeS = re.compile(r'\d\d:\d\d:\d\d.\d\d')
    
    Time = re.findall(TimeS, data)
    df['TimeStamp'] = Time
                   
    ACC_X = re.findall('ACC_X:\s.+',data)
    lis = []
    for term in ACC_X:
        term = re.search('(ACC_X:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['ACC_X'] = lis
    
    ACC_Y = re.findall('ACC_Y:\s.+',data)
    lis = []
    for term in ACC_Y:
        term = re.search('(ACC_Y:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['ACC_Y'] = lis
    
    ACC_Z = re.findall('ACC_Z:\s.+',data)
    lis = []
    for term in ACC_Z:
        term = re.search('(ACC_Z:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['ACC_Z'] = lis
    
    GYR_X = re.findall('GYR_X:\s.+',data)
    lis = []
    for term in GYR_X:
        term = re.search('(GYR_X:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['GYR_X'] = lis
    
    GYR_Y = re.findall('GYR_Y:\s.+',data)
    lis = []
    for term in GYR_Y:
        term = re.search('(GYR_Y:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['GYR_Y'] = lis   
    
    GYR_Z = re.findall('GYR_Z:\s.+',data)
    lis = []
    for term in GYR_Z:
        term = re.search('(GYR_Z:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['GYR_Z'] = lis
    
    MAG_X = re.findall('MAG_X:\s.+',data)
    lis = []
    for term in MAG_X:
        term = re.search('(MAG_X:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['MAG_X'] = lis
    
    MAG_Y = re.findall('MAG_Y:\s.+',data)
    lis = []
    for term in MAG_Y:
        term = re.search('(MAG_Y:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['MAG_Y'] = lis
    
    MAG_Z = re.findall('MAG_Z:\s.+',data)
    lis = []
    for term in MAG_Z:
        term = re.search('(MAG_Z:\s)(.+)',term).group(2)
        term = int(term)
        lis.append(term)
    df['MAG_Z'] = lis
    
    PRESS = re.findall('PRESS:\s.+',data)
    lis = []
    for term in PRESS:
        term = re.search('(PRESS:\s)(.+)',term).group(2)
        term = float(term)
        lis.append(term)
    df['PRESS'] = lis
            
    return df
    
```

```python
def get_px(df):
    vx= []
    t = 0.05
    x = df.loc[:,['ACC_X']].ACC_X
    for n in range(0,18):
        if n == 0:
            vx.append(0)
            n += 1
        else:
            vx.append(vx[n-1] + 0.5 *(x[n-1]+x[n])*t)
            n += 1
    px = get_p(vx)
            
    return px

        
def get_py(df):
    vy= []
    t = 0.05
    y = df.loc[:,['ACC_Y']].ACC_Y
    for n in range(0,18):
        if n == 0:
            vy.append(0)
            n += 1
        else:
            vy.append(vy[n-1] + 0.5 *(y[n-1]+y[n])*t)
            n += 1
    
    py = get_p(vy)
    
    return py
        
def get_pz(df):
    vz= []
    t = 0.05
    z = df.loc[:,['ACC_Z']].ACC_Z
    for n in range(0,18):
        if n == 0:
            vz.append(0)
            n += 1
        else:
            vz.append(v[n-1] + 0.5 *(z[n-1]+z[n])*t)
            n += 1
            
    pz = get_p(vz)
    
    return pz
```

```python
x1 = get_px(data1)
y1 = get_py(data1)
z1 = get_pz(data1)

fig = plt.figure()
axes3d = Axes3D(fig)

trace1 = axes3d.plot(x1, y1, z1, c = 'r')

plt.show()
```
